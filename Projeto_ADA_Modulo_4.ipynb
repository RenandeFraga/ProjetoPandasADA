<h4 style='color:green'> Análise: </h4>

```python
data_pedidos = tabela_completa[['codigo_pedido', 'data_pedido']]
```

```python
data_pedidos
```

```python
somente_data_pedidos = data_pedidos['data_pedido'].tolist()
somente_horarios = []
for h in range(0, len(somente_data_pedidos)):
    somente_horarios.append(somente_data_pedidos[h][11:])
```

```python
horas = []
for h in range(0, len(somente_horarios)):
    horas.append(somente_horarios[h][:2])
```

```python
x = np.array(horas)
x = x.astype(int)
unique, counts = np.unique(x, return_counts=True)
horarios_compras = pd.DataFrame({'horario' : unique, 'total_pedidos' : counts})
horarios_compras
```

```python
#dia da semana
so_data = data_pedidos['data_pedido'].to_list()
datas_pedidos = []
for h in range(0, len(so_data)):
    datas_pedidos.append(so_data[h][:10])
datas_compras = pd.DataFrame({'pedido' : data_pedidos['codigo_pedido'], 'data' : datas_pedidos})
datas_compras['data'] = pd.to_datetime(datas_compras['data']) 
```

```python
datas_compras.dtypes
```

```python

datas_compras['dia_semana'] = datas_compras['data'].dt.dayofweek
dict_dias_semana = {0:'seg', 1:'ter', 2:'qua', 3:'qui', 4:'sex', 5:'sab', 6:'dom'}
datas_compras['dia_semana_s'] = datas_compras['dia_semana'].map(dict_dias_semana)
datas_compras['mes'] = datas_compras['data'].dt.month
dict_month = {1 : 'janeiro', 2 : 'fevereiro', 3 : 'março', 4 : 'abril', 5 : 'maio', 6 : 'junho', 7 : 'julho', 8 : 'agosto', 9 : 'setembro', 10 : 'outubro', 11 : 'novembro', 12 : 'dezembro' }
datas_compras['mes_s'] = datas_compras['mes'].map(dict_month)
datas_compras
```

```python
# periodo

periodos_horas = [0, 6, 12, 18, 23]
labels = ['madrugada', 'manhã', 'tarde', 'noite']
horarios_compras['periodo_dia'] = pd.cut(x=horarios_compras['horario'], bins=periodos_horas, labels=labels)
horarios_compras.fillna('madrugada', inplace=True)
```
